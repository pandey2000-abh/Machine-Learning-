```python
# divide the dataset into independent and dependent feature  x is independent
```

```python
x = df.iloc[:,:-1]  # for remove last col
```

```python
y = df.iloc[:,-1]
```

```python
#first step to devide in test and test data-> due to data leakage 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state = 32)
```

```python
x_train.head()
```

```python
x.shape
```

```python
x_train.shape,x_test.shape
```

```python
# feature scaling
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() # standardization form z_score formula
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
```

```python
x_train
```

```python
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
# model train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
# prediction
```

```python
y_pred=regression.predict(x_test)
```

```python
# MSE,MAE,RMSE
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
import numpy as np
mse = mean_squared_error(y_test,y_pred)
print(mse)
mae = mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
# for accuracy r2 and r square
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)
```

```python
print(score)
```

```python
# display adjusted R-squared
```

```python
1-(1-score)*(len(y)-1)/(len(y)-x.shape[1]-1)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(x_train,y_train)
```

```python
y_pred = ridge.predict(x_test)
```

```python
import numpy as np
mse = mean_squared_error(y_test,y_pred)
print(mse)
mae = mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=20.0)
lasso.fit(x_train,y_train)
```

```python
y_pred = lasso.predict(x_test)
```

```python
import numpy as np
mse = mean_squared_error(y_test,y_pred)
print(mse)
mae = mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
from sklearn.linear_model import ElasticNet
elasticnet = ElasticNet(alpha=20.0)
elasticnet.fit(x_train,y_train)
```

```python
y_pred = elasticnet.predict(x_test)
```
